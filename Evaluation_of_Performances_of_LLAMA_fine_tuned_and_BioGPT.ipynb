```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from transformers import AutoModelForCausalLM, AutoTokenizer, GenerationConfig, pipeline
```

```python
model_id = "Narrativaai/BioGPT-Large-finetuned-chatdoctor"

tokenizer = AutoTokenizer.from_pretrained("microsoft/BioGPT-Large")

model = AutoModelForCausalLM.from_pretrained(model_id)
```

```python
def answer_question(prompt,temperature=0.1,top_p=0.75,top_k=40,num_beams=2,**kwargs):
    inputs = tokenizer(prompt, return_tensors="pt")
    input_ids = inputs["input_ids"]
    attention_mask = inputs["attention_mask"]
    generation_config = GenerationConfig(
        temperature=temperature,
        top_p=top_p,
        top_k=top_k,
        num_beams=num_beams,
        **kwargs,
    )
    with torch.no_grad():
        generation_output = model.generate(
            input_ids=input_ids,
            attention_mask=attention_mask,
            generation_config=generation_config,
            return_dict_in_generate=True,
            output_scores=True,
            max_new_tokens=600,
            eos_token_id=tokenizer.eos_token_id

        )
    s = generation_output.sequences[0]
    output = tokenizer.decode(s, skip_special_tokens=True)
    return output.split(" Response:")[1]
```

```python
def answer_BioGPT(prompt):
    prompt_template = """
    Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

    ### Instruction:
    As a nutritionist, provide dietary advice based on the patient's description.

    ### Input:
    {}
    ### Response:
    """.format(prompt)
    return answer_question(prompt_template)
```

```python
import os
import torch
from datasets import load_dataset
from transformers import (
    AutoModelForCausalLM,
    AutoTokenizer,
    BitsAndBytesConfig,
    HfArgumentParser,
    TrainingArguments,
    pipeline,
    logging,
)
from peft import LoraConfig, PeftModel
from trl import SFTTrainer
```

```python
model_name = "NousResearch/Llama-2-7b-chat-hf"

# The instruction dataset to use
dataset_name = "pubmed_qa"

# Fine-tuned model name
new_model = "Llama-2-7b-medQnA-finetune"

################################################################################
# QLoRA parameters
################################################################################

# LoRA attention dimension
lora_r = 64

# Alpha parameter for LoRA scaling
lora_alpha = 16

# Dropout probability for LoRA layers
lora_dropout = 0.1

################################################################################
# bitsandbytes parameters
################################################################################

# Activate 4-bit precision base model loading
use_4bit = True

# Compute dtype for 4-bit base models
bnb_4bit_compute_dtype = "float16"

# Quantization type (fp4 or nf4)
bnb_4bit_quant_type = "nf4"

# Activate nested quantization for 4-bit base models (double quantization)
use_nested_quant = False

################################################################################
# TrainingArguments parameters
################################################################################

# Output directory where the model predictions and checkpoints will be stored
output_dir = "./results"

# Number of training epochs
num_train_epochs = 1

# Enable fp16/bf16 training , it enables for A100 GPU
fp16 = False
bf16 = False

# Batch size per GPU for training
per_device_train_batch_size = 4

# Batch size per GPU for evaluation
per_device_eval_batch_size = 4

# Number of update steps to accumulate the gradients for
gradient_accumulation_steps = 1

# Enable gradient checkpointing
gradient_checkpointing = True

# Maximum gradient normal (gradient clipping)
max_grad_norm = 0.3

# Initial learning rate (AdamW optimizer)
# Weight decay to apply to all layers except bias/LayerNorm weights
weight_decay = 0.001

# Optimizer to use
optim = "paged_adamw_32bit"

# Learning rate schedule
lr_scheduler_type = "cosine"

learning_rate = 2e-4

# Number of training steps (overrides num_train_epochs)
max_steps = -1

# Ratio of steps for a linear warmup (from 0 to learning rate)
warmup_ratio = 0.03

# Group sequences into batches with same length
# Saves memory and speeds up training considerably
group_by_length = True

# Save checkpoint every X updates steps
save_steps = 0

# Log every X updates steps
logging_steps = 25

# SFT parameters
################################################################################

# Maximum sequence length to use
max_seq_length = None

# Pack multiple short examples in the same input sequence to increase efficiency
packing = False

device_map = {"": 0}
```

```python
dataset = load_dataset(dataset_name, "pqa_artificial")  # Remove the extra parameter

# Load tokenizer and model with QLoRA configuration
compute_dtype = getattr(torch, bnb_4bit_compute_dtype)

bnb_config = BitsAndBytesConfig(
    load_in_4bit=use_4bit,
    bnb_4bit_quant_type=bnb_4bit_quant_type,
    bnb_4bit_compute_dtype=compute_dtype,
    bnb_4bit_use_double_quant=use_nested_quant,
)
```

```python
# Load base
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map=device_map
)
model.config.use_cache = False
model.config.pretraining_tp = 1

# Load LLaMA tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"  # Fix weird overflow issue with fp16
```

```python
peft_config = LoraConfig(
    lora_alpha=lora_alpha,
    lora_dropout=lora_dropout,
    r=lora_r,
    bias="none",
    task_type="CAUSAL_LM",
)

# Set training parameters
training_arguments = TrainingArguments(
    output_dir=output_dir,
    num_train_epochs=num_train_epochs,
    per_device_train_batch_size=per_device_train_batch_size,
    gradient_accumulation_steps=gradient_accumulation_steps,
    optim=optim,
    save_steps=save_steps,
    logging_steps=logging_steps,
    learning_rate=learning_rate,
    weight_decay=weight_decay,
    fp16=fp16,
    bf16=bf16,
    max_grad_norm=max_grad_norm,
    max_steps=max_steps,
    warmup_ratio=warmup_ratio,
    group_by_length=group_by_length,
    lr_scheduler_type=lr_scheduler_type,
    report_to="tensorboard"
)
```

```python
# Selecting the "train" split
train_dataset = dataset["train"]

# Define the function to merge question and answer
def merge_question_and_answer(example):
    return {"text": f"Question: {example['question']} Context: {example['long_answer']}"}

# Apply the function to the selected split
train_dataset = train_dataset.map(merge_question_and_answer)

# Set up SFTTrainer with the selected split
trainer = SFTTrainer(
    model=model,
    train_dataset=train_dataset,  # Use the selected split
    peft_config=peft_config,
    dataset_text_field="text",
    max_seq_length=max_seq_length,
    tokenizer=tokenizer,
    args=training_arguments,
    packing=packing,
)

# Train the model
trainer.train()
```

```python
import os
import nltk
from nltk.translate.bleu_score import sentence_bleu
from rouge import Rouge
from bert_score import score
import Levenshtein
nltk.download('punkt')
```

```python
from transformers import pipeline
pipe = pipeline(task="text-generation", model=model, tokenizer=tokenizer, max_length=300)
def generate_answer(prompt):
    result = pipe(prompt)
    generated_text = result[0]['generated_text']
    answer_index = generated_text.find("Answer:")
    if answer_index != -1:
        answer_text = generated_text[answer_index + len("Answer:"):]
        return answer_text.strip()
    else:
        return "No answer found."
```

```python
from collections import Counter
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
from nltk.lm import Vocabulary

def compute_diversity(generated_texts):
    num_unique = len(set(generated_texts))
    diversity = num_unique / len(generated_texts)
    return diversity

def compute_novelty(generated_texts, reference_texts):
    reference_vocab = Vocabulary(word_tokenize(' '.join(reference_texts)))
    generated_vocab = Vocabulary(word_tokenize(' '.join(generated_texts)))
    novel_tokens = set(generated_vocab) - set(reference_vocab)
    novelty = len(novel_tokens) / len(generated_vocab)
    return novelty

diversity_biogpt = []
diversity_llama = []


novelty_biogpt = []
novelty_llama = []


bleu_bioGPT = []
bleu_llama = []

rouge_bioGPT_scores = []
rouge_llama_scores = []

bscore_bioGPT = []
bscore_llama = []

lev_bioGPT = []
lev_llama = []

train_dataset = dataset['train']
count = 0

rouge = Rouge()

for example in train_dataset:
  if count == 20:
    break

  input_text = example['input']
  answer = example['output']

  biogpt_answer = answer_BioGPT(input_text)
  llama_answer = generate_answer(input_text)

  token_actual = nltk.word_tokenize(answer)
  token_input = nltk.word_tokenize(input_text)

  bleu_bioGPT.append(sentence_bleu([token_actual], token_input))
  bleu_llama.append(sentence_bleu([token_actual], nltk.word_tokenize(llama_answer)))

  rouge_bioGPT_scores.append(rouge.get_scores(answer, biogpt_answer))
  rouge_llama_scores.append(rouge.get_scores(answer, llama_answer))

  _, _, F1_bioGPT = score([biogpt_answer], [answer], lang='en', verbose=True)
  _, _, F1_llama = score([llama_answer], [answer], lang='en', verbose=True)
  bscore_bioGPT.append(F1_bioGPT.item())
  bscore_llama.append(F1_llama.item())

  lev_bioGPT.append(Levenshtein.distance(answer, biogpt_answer))
  lev_llama.append(Levenshtein.distance(answer, llama_answer))

  novelty_llama.append(compute_novelty(llama_answer, answer))
  novelty_biogpt.append(compute_novelty(biogpt_answer, answer))

  diversity_llama.append(compute_diversity(llama_answer))
  diversity_biogpt.append(compute_diversity(biogpt_answer))
  count += 1
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))

# Plot BLEU scores
plt.subplot(2, 3, 1)
plt.plot(bleu_bioGPT, label='BioGPT')
plt.plot(bleu_llama, label='LLAMA')
plt.xlabel('Example')
plt.ylabel('BLEU Score')
plt.title('BLEU Score Comparison')
plt.legend()

# # Plot ROUGE scores
# plt.subplot(2, 3, 2)
# plt.plot([score['rouge-l']['f'] for score in rouge_bioGPT_scores], label='BioGPT')
# plt.plot([score['rouge-l']['f'] for score in rouge_llama_scores], label='LLAMA')
# plt.xlabel('Example')
# plt.ylabel('ROUGE-L Score')
# plt.title('ROUGE-L Score Comparison')
# plt.legend()

# Plot BERTScores
plt.subplot(2, 3, 3)
plt.plot(bscore_bioGPT, label='BioGPT')
plt.plot(bscore_llama, label='LLAMA')
plt.xlabel('Example')
plt.ylabel('BERTScore')
plt.title('BERTScore Comparison')
plt.legend()

# Plot Levenshtein Distance
plt.subplot(2, 3, 4)
plt.plot(lev_bioGPT, label='BioGPT')
plt.plot(lev_llama, label='LLAMA')
plt.xlabel('Example')
plt.ylabel('Levenshtein Distance')
plt.title('Levenshtein Distance Comparison')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
def mean(lst):
    return sum(lst) / len(lst)

mean_bleu_biogpt = mean(bleu_bioGPT)
mean_bleu_llama = mean(bleu_llama)

mean_novelty_biogpt = mean(novelty_biogpt)
mean_novelty_llama = mean(novelty_llama)
```
